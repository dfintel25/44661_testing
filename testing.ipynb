```python
data = pd.read_csv (
    r"C:\Users\derek\OneDrive\Documents\44661_SQL Testing\country.csv",
    encoding='latin1',
    dtype={'zip_code': str}
)
data.head()
```

```python
# Filter to only official languages
official_langs = data[data['IsOfficial'] == 'T']

# Count official languages per country
lang_counts = official_langs.groupby('CountryCode')['Language'].count().sort_values(ascending=False)
```

```python
# Plot
plt.figure(figsize=(10, 6))
lang_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Official Languages per Country')
plt.xlabel('Country Code')
plt.ylabel('Number of Official Languages')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
